```python
import nltk
import pandas as pd
import xlrd
import datetime
import ntpath
import os
nltk.download('vader_lexicon')
```

```python
# Give the location of the file 
location = ("C:\Users\Sephali.panda\Desktop\New bot\New\1592258.xlsx")
filename = ntpath.basename(location)

# To open Workbook 
wb = xlrd.open_workbook(location) 
sheet = wb.sheet_by_index(0) 
rows = sheet.nrows

dataset = []
for i in range (0,rows):
    speech_1 = str(sheet.cell(i, 0).value)
    dataset.append(speech_1)
```

```python
def nltk_sentiment(sentence):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    nltk_sentiment = SentimentIntensityAnalyzer()
    score = nltk_sentiment.polarity_scores(sentence)
    return score
```

```python


nltk_results = [nltk_sentiment(row) for row in dataset]
results_df = pd.DataFrame(nltk_results)
text_df = pd.DataFrame(dataset, columns = ['text'])
nltk_df = text_df.join(results_df)
```

```python
nltk_df.head(300)
```

```python
negative_comments_df = nltk_df[nltk_df['compound'] < 0]
positive_comments_df = nltk_df[nltk_df['compound'] > 0]
```
